```python
path = '/content/drive/MyDrive/gita_processed2.csv'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
dataset = []
```

```python
for _, row in df.iterrows():
  prompt = f"What does Chapter {row['Chapter']}, Verse {row['Verse']} of the Bhagavad Gita mean?"
  completion = row['Swami Sivananda']
  dataset.append({'text':f'<s>[INST] {prompt} [/INST] {completion} </s>'})
```

```python
print(dataset[0])
```

```python
concepts.head()
```

```python
for _, row in concepts.iterrows():
  prompt = f"What does Bhagwad Gita say about {row['Concept']}?"
  completion = row['English']
  dataset.append({'text':f'<s>[INST] {prompt} [/INST] {completion} </s>'})
```

```python
for _, row in df.iterrows():
  prompt = f"I came across this shloka of Bhagwad Gita, {row['Sanskrit ']}. What does this mean?"
  completion = row['Swami Sivananda']
  dataset.append({'text':f'<s>[INST] {prompt} [/INST] {completion} </s>'})
```
